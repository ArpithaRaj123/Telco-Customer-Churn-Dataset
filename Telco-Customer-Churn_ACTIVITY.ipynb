```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset=pd.read_excel("Telco-Customer-Churn.xlsx")
```

```python
dataset
```

# 1) Compare churn count with respect to gender.

```python
plt.figure(figsize=(8,8))
sns.countplot(x='Churn',data=dataset,hue='gender')
plt.title('COMPARISON OF CHURN COUNT VS GENDER',fontsize=15)
plt.grid()
```

# 2) Find out how many female senior citizens there in the dataset

```python
females_senior_citizen = dataset[(dataset['gender']=='Female') & (dataset['SeniorCitizen']==1)]
```

```python
females_senior_citizen
```

```python
females_senior_citizen.groupby(['SeniorCitizen','gender'])['gender'].count()
```

# 3) Compare 'tenure' with 'Total Charges'¶

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x=dataset['tenure'],y=dataset['TotalCharges'],data=dataset,hue='SeniorCitizen')
plt.title('PLOT OF TENURE VS TOTALCHARGES',fontsize=15)
plt.grid()
```

```python
From the above-plotted graph, we can conclude that total charges increases as per tenure value increases, and non-senior citizens use more tenure conditions compared to senior citizens. 
```

# 4) Find out which contract preffered by the senior citizen. 

```python
senior_citiz=dataset[(dataset['SeniorCitizen']==1)]
```

```python
senior_citiz
```

```python
contract_prefered=senior_citiz.groupby(['SeniorCitizen','Contract']).sum()
```

```python
contract_prefered
```

```python
#The Total Charges for "MONTH-TO-MONTH is highest compared to other contract methods from the dataset"
```

```python
#Visualization of the most prefered contract by senior citizens
plt.figure(figsize=(8,8))
sns.countplot(x='Contract',data=dataset)
plt.title('PLOT OF CONTRACTS',fontsize=15)
plt.grid()
```

```python
#From the above plotted graph we can conclude that MONTH-TO-MONTH contract is prefred by most of the senior ctizens.
```
